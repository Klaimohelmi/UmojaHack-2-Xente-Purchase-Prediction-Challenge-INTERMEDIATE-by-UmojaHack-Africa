```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
sub = pd.read_csv('sampleSubmission.csv')
submission=sub.copy()
```

```python
sub = pd.read_csv('sampleSubmission.csv')

# Make sure train is in the folder where this notebook is running
train = pd.read_csv('Train.csv')

pid = pd.read_csv('PID_Categories.csv')
train['d']=train['date'].apply(lambda x: x.split(' ')[0])
```

```python
sub['acc']=sub['Account X date X PID'].apply(lambda x: x.split('X')[0]).astype(float)
```

```python
sub['d']=sub['Account X date X PID'].apply(lambda x: x.split('X')[1])
```

```python
sub['PID']=sub['Account X date X PID'].apply(lambda x: x.split(' X ')[2])
```

```python
import datetime
a=list(sub['acc'].unique())
b=pd.date_range(end = datetime.datetime.strptime("22-02-2020", "%d-%m-%Y"), periods = 60).to_pydatetime().tolist()
c=list(sub['PID'].unique())



index = pd.MultiIndex.from_product([a,b, c], names = ["acc", 'd',"PID"])

new=pd.DataFrame(index = index).reset_index()
```

```python
group=train[train['d']>'2019-12-5']
```

```python
group=group.groupby(['acc','PID','d']).count()
group.columns=['Prediction']
group=group.reset_index()
```

```python
group['d']=pd.to_datetime(group['d'])
```

```python
sub['d']=pd.to_datetime(sub['d'])
```

```python
new=new.merge(group,on=['acc','PID','d'],how='left')
```

```python
new.fillna(0,inplace=True)
```

```python
data_pid=train.groupby('PID').count()[['date']]
data_pid.columns=['data_pid']
new=new.merge(data_pid,on=['PID'],how='left')
sub=sub.merge(data_pid,on=['PID'],how='left')
data_acc=train.groupby('acc').count()['date']
data_acc.columns=[['data_acc']]
new=new.merge(data_acc,on=['acc'],how='left')
sub=sub.merge(data_acc,on=['acc'],how='left')
#from sklearn import preprocessing
#le = preprocessing.LabelEncoder()
#le.fit(pid['Category'])

#pid['Category']=le.transform(pid['Category'])
```

```python
#new=new.merge(pid,on=['PID'],how='left')
#sub=sub.merge(pid,on=['PID'],how='left')
```

```python
def pred(col):
    if col>0:
        return 1
    else:
        return 0
new['Prediction']=new['Prediction'].apply(pred)
```

```python
new['dow']=new['d'].dt.dayofweek
sub['dow']=sub['d'].dt.dayofweek
```

```python
sub['d']=pd.to_datetime(sub['d'])
new['dom']=new['d'].dt.month
sub['dom']=sub['d'].dt.month
```

```python
new[new['Prediction']>0]
```

```python
ids_new=pd.get_dummies(new['PID'])
ids_sub=pd.get_dummies(sub['PID'])
```

```python
new=pd.concat([new,ids_new],axis=1)
sub=pd.concat([sub,ids_sub],axis=1)
```

```python
ids_new=pd.get_dummies(new['acc'])
ids_sub=pd.get_dummies(sub['acc'])
new=pd.concat([new,ids_new],axis=1)
sub=pd.concat([sub,ids_sub],axis=1)
```

```python
drop_sub=['d','acc','Account X date X PID','PID','Prediction','dom']
drop_new=['d','acc','PID','dom']
```

```python
sub=sub.drop(columns=drop_sub)
new=new.drop(columns=drop_new)
```

```python

X,y=new.drop(columns='Prediction',axis=1),new['Prediction']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=42)
```

```python

from catboost import CatBoostClassifier

# Create the model
model = CatBoostClassifier()
```

```python
model.fit(X_train,y_train)

from sklearn.metrics import f1_score
f1_score(y_test, model.predict(X_test),average='binary')
```

```python
model.fit(X,y)
```

```python
prediction=model.predict(sub)
```

```python
submission['Prediction']=prediction
```
